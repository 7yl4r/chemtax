```python
import ipywidgets as widgets
from ipywidgets import Layout

print("""
STEP 1 : Upload data for analysis.
-----------------------------------------------------------------
""")
print("""
1.1) upload the samples pigment data table. This can be:
    * a .csv file with one sample per row, one pigment per column
    * Not Yet Implemented: a SeaBASS file? (also includes lat, lon, timestamp)
""")
pigment_samples_uploader = widgets.FileUpload(
    accept='.csv',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False,  # True to accept multiple files upload else False
    description="Upload Pigments per Sample .csv",
    layout=Layout(width='100%', height='50px')
)
display(pigment_samples_uploader)

print("""
1.2) upload the list of taxa present. This should be a comma-separated list of taxa names.
""")
taxa_list_uploader = widgets.FileUpload(
    accept='.csv',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False,  # True to accept multiple files upload else False
    description="Upload taxa list file.",
    layout=Layout(width='100%', height='50px')
)
display(taxa_list_uploader)
```
